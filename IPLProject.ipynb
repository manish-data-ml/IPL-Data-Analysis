## Analysis

### Q1: Which team won the most matches?

```python
matchs_wins=df['match_winner'].value_counts()
matchs_wins
```

```python
sns.barplot(y=matchs_wins.index,x=matchs_wins.values,palette='viridis')
plt.title('Most Matchs Win By Teams')
plt.xlabel("Match_Winner")
```

From the analysis, Gujarat team has the highest number of match wins. This suggests that Gujarat was the most dominant team in the given dataset.

## Q2.Toss Decisions Trands

```python
df.head()
sns.countplot(x=df['toss_decision'],palette='viridis')
plt.title('Toss_Decision')
```

From the analysis, most teams prefer to choose fielding after winning the toss. This suggests that teams believe chasing is easier than setting a target in IPL matches.

## Q3.Won By?

```python
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=count*100/df.shape[0]
percentage.round(2)
```

```python
sns.countplot(x=df['won_by'])   #wonby
```

## Q4. Player of the Match Explanation

```python
top_player=df['player_of_the_match'].value_counts().head(10)
sns.barplot(x=top_player.values,y=top_player.index,palette='viridis')
plt.title('Top 10 player of the match Award')
```

The analysis shows that some players have won the Player of the Match award multiple times. This indicates their consistent performance and impact in matches.

## Q5. Top Scorer 

```python
TopScorer=df['top_scorer'].value_counts().head()
TopScorer
sns.barplot(x=TopScorer.values,y=TopScorer.index,palette='viridis')
```

From the analysis, Jos Buttler appears as the top scorer with the highest number of high-scoring performances. Other players like Quinton de Kock, Liam Livingstone, Shubman Gill, and KL Rahul also show strong batting performance in the dataset.

## Q6. Best Bowling?

```python
Best_bowling=df['best_bowling'].value_counts().head(10)
Best_bowling
sns.barplot(x=Best_bowling.values,y=Best_bowling.index, palette='coolwarm')
```

From the analysis, Yuzvendra Chahal has the highest number of best bowling performances, indicating his strong impact as a bowler. Other bowlers like Rashid Khan, T Natarajan, and Kagiso Rabada also show consistent performance in the dataset.
